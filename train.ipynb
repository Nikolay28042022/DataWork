Некоторые столбцы имеют очень высокий процент пропущенных данных, такие как PoolQC, MiscFeature, Alley и Fence (более 80%), что делает их малозначимыми для модели. Пропуски в таких столбцах можно удалить или игнорировать. Остальные столбцы содержат меньше пропусков, и их можно заполнить соответствующими статистическими значениями.

```python
# 1. Удаление столбцов с более чем 80% пропусков
columns_to_drop = ['PoolQC', 'MiscFeature', 'Alley', 'Fence']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df
```

```python
# 2. Заполнение медианой для числовых столбцов с пропусками
numerical_columns_with_missing = ['LotFrontage', 'GarageYrBlt', 'MasVnrArea']
for col in numerical_columns_with_missing:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
# 3. Заполнение модой для категориальных столбцов с пропусками
categorical_columns_with_missing = ['GarageQual', 'GarageFinish', 'GarageCond', 'GarageType', 'BsmtCond',
                                    'BsmtQual', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrType',
                                    'Electrical']
for col in categorical_columns_with_missing:
    df[col].fillna(df[col].mode()[0], inplace=True) #   Метод fillna заменяет все значения NaN (пропущенные значения) на указанное значение
```

```python
# 4. Проверка и удаление дубликатов
duplicates = df.duplicated().sum()

# Удаление дубликатов, если они есть
df.drop_duplicates(inplace=True)
```

```python
# 5. Анализ и преобразование типов данных
# Рассмотрим объектные (категориальные) столбцы
object_columns = df.select_dtypes(include=['object']).columns

# Для дальнейшего анализа выполню one-hot encoding для категориальных переменных
df_encoded = pd.get_dummies(df, columns=object_columns, drop_first=True)

# Проверим результат
df_encoded.info(), duplicates  # Посмотрим на инфо после преобразований, а также количество дубликатов
```

```python
#    Создадим новые признаки.
# Создадим признак 'TotalArea' (общая площадь), объединяющий GrLivArea (жилая площадь) и TotalBsmtSF (площадь подвала).
df_encoded['TotalArea'] = df_encoded['GrLivArea'] + df_encoded['TotalBsmtSF']

# Добавим признак возраста дома 'HouseAge', вычислив его как разницу между текущим годом и годом постройки.
df_encoded['HouseAge'] = df_encoded['YrSold'] - df_encoded['YearBuilt']

```
